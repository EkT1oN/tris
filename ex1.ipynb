```python
from apriori_python import apriori
import numpy as np
```

```python
itemArray = [
    '0100110011101101',
    '1011001101110010',
    '0001110110111001',
    '0110011001100111',
    '1101010101010110',
    '0010111010101011',
    '1101000011001111',
    '0011011011101010'
]
oneClass = [
    'голова круглая',
    'уши оттопыренные',
    'нос круглый',
    'глаза круглые',
    'лоб с морщинами',
    'носогубная складка есть',
    'губы толстые',
    'волосы есть',
    'усы есть',
    'борода есть',
    'очки есть',
    'родинка на щеке есть',
    'бабочка есть',
    'брови подняты кверху',
    'серьга есть',
    'курительная трубка есть'
]
zeroClass = [
    'голова овальная',
    'уши прижатые',
    'нос длинный',
    'глаза узкие',
    'лоб с без морщин',
    'носогубная складка нет',
    'губы тонкие',
    'волосы нет',
    'усы нет',
    'борода нет',
    'очки нет',
    'родинка на щеке нет',
    'бабочка нет',
    'брови опущены книзу',
    'серьга нет',
    'курительная трубка нет'
]
```

```python
transactions = []
for string in itemArray:
    tmp = []
    for i in range(len(string)):
        if string[i] == '0':
            tmp.append(zeroClass[i])
        if string[i] == '1':
            tmp.append(oneClass[i])
    transactions.append(tmp)
print(f'{transactions}')
```

```python
freqItemSet, rules = apriori(transactions, minSup=0.5, minConf=0.5)
print(rules)
```
