# Импортирование библиотек

```python
from efficient_apriori import apriori
from tabulate import tabulate
import numpy as np
```

# Загрузка входных данных

```python
itemArray = [
    '0100110011101101',
    '1011001101110010',
    '0001110110111001',
    '0110011001100111',
    '1101010101010110',
    '0010111010101011',
    '1101000011001111',
    '0011011011101010'
]
oneClass = [
    'голова круглая',
    'уши оттопыренные',
    'нос круглый',
    'глаза круглые',
    'лоб с морщинами',
    'носогубная складка есть',
    'губы толстые',
    'волосы есть',
    'усы есть',
    'борода есть',
    'очки есть',
    'родинка на щеке есть',
    'бабочка есть',
    'брови подняты кверху',
    'серьга есть',
    'курительная трубка есть'
]
zeroClass = [
    'голова овальная',
    'уши прижатые',
    'нос длинный',
    'глаза узкие',
    'лоб с без морщин',
    'носогубная складка нет',
    'губы тонкие',
    'волосы нет',
    'усы нет',
    'борода нет',
    'очки нет',
    'родинка на щеке нет',
    'бабочка нет',
    'брови опущены книзу',
    'серьга нет',
    'курительная трубка нет'
]
```

# Составление "характеристик" для каждого лица

```python
transactions = []
for string in itemArray:
    tmp = []
    for i in range(len(string)):
        if string[i] == '0':
            tmp.append(zeroClass[i])
        if string[i] == '1':
            tmp.append(oneClass[i])
    transactions.append(tmp)
# i = 1
# for elem in transactions:
#     print(f'{i}: {elem}')
transactions[0]
    
```

# Поиск ассоциативных правил

```python
itemsets, rules = apriori(transactions, min_support=0.5, min_confidence=1)
```

# Вывод результатов

```python
to_str = lambda rule: ', '.join(list(rule))
results = [
    [
        f'{to_str(rule.lhs)} -> {to_str(rule.rhs)}',
        rule.confidence,
        rule.support,
        rule.lift,
        rule.conviction,
    ]
    for rule in sorted(rules, key=lambda rule: rule.lift, reverse=True) ] # сортировка по Lift
print(tabulate(results, headers=['Rule', 'Confidence', 'Support', 'Lift', 'Conviction']))
```
